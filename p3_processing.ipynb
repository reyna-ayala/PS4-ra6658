```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x128 = np.linspace(0,1,128); y128=x128
x256 = np.linspace(0,1,256); y256=x256
x512 = np.linspace(0,1,512); y512=x512

a = 1
b = 2
dx = x128[1] - x128[0]
dt = dx / (a+b) * 0.25
t = range(len(err_LW_128)) * dt

ut1_LW_128 = pd.read_csv('ut1_LW_N128.csv')
ut1_LW_128 = np.array(ut1_LW_128)[:,1:]

ut10_LW_128 = pd.read_csv('ut10_LW_N128.csv')
ut10_LW_128 = np.array(ut10_LW_128)[:,1:]

err_LW_128 = pd.read_csv('err_LW_N128.csv')
err_LW_128 = np.array(err_LW_128)[:,1:]

plt.figure()
plt.contourf(x128,y128,ut1_LW_128)
plt.colorbar(label='u')
plt.xlabel('x')
plt.ylabel('y')
plt.title('LW Contour plot of u: t=1, N=128')

plt.figure()
plt.contourf(x128,y128,ut10_LW_128)
plt.colorbar(label='u')
plt.xlabel('x')
plt.ylabel('y')
plt.title('LW Contour plot of u: t=10, N=128')

plt.figure()
plt.plot(t,err_LW_128)
plt.xlabel('t', size=14)
plt.ylabel(r'$L_2$ norm', size=14)

plt.show()
```
